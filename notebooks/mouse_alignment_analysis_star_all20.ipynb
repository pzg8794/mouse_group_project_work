# Mouse Alignment Analysis — STAR All 20 (`SRP618841`)

This is the team-share version of the `mouse_new` alignment walkthrough. It summarizes the shared `SRP618841` STAR run and focuses on three practical questions:

1. Are the alignment metrics strong enough to proceed to DE?
2. Do the remaining GC-WARN samples look different at the alignment stage?
3. Is there any major platform split left to worry about in this dataset?


```python
from pathlib import Path
import pandas as pd
from IPython.display import Image, display

ROOT = Path('..').resolve()
DATA = ROOT / 'data' / 'alignment_analysis_star_all20'
TABLES = DATA / 'tables'
FIGS = DATA / 'figures'
```

## Shared alignment inputs

These artifacts were exported from the shared `SRP618841` alignment run on `sequoia` and copied into this repo so the team can inspect them without using the full private working tree.

## Main takeaways

- The `SRP618841` alignment layer is strong enough to proceed to DE.
